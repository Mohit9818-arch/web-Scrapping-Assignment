# Data Visualization using Matplotlib

```python
plt.plot()for plotting line chart similarily in place of plot other functions are used
for plotting. All plotting functions require data and it is provided in the function
through parameters.plot.xlabel,plt.ylabel for labelling x and y-axis respectively.
```

```python
#from matplotlib import pyplot as plt
import matplotlib.pyplot as plt
import numpy as np
```

```python
a=np.array([1,2,3,4,5,6,7])
b=np.array([1,2,3,4,5,6,7])
```

```python
plt.title("Linear Line")
plt.xlabel("a")
plt.ylabel("b")
plt.plot(a,b)
plt.show()
```

```python
g=np.array([3,6,44,12,67,45,55])
h=np.array([22,4,5,6,78,22,12])
plt.xlabel('g')
plt.ylabel('h')
plt.plot(g,h)
plt.show()
```

```python
plt.plot(g,h,'--g')
plt.show()
```

```python
x=[5,8,10,2,6,12]
y=[12,16,6,10,14,6]
plt.bar(x,y)
plt.title('Bar Graph')
plt.ylabel('Y axis')
plt.xlabel('X axis')
plt.show()
```

```python
plt.bar(x,y,color = 'm', align= 'center')
plt.title('Bar graph')
plt.ylabel('Y axis')
plt.xlabel('X axis')
plt.show()
```

Histogram

```python
A histogram takes in a series of data and divides the data into a number of bins.
It then plots the frequency data points in each bin(i.e the interval of points). It
is useful in understanding the count of data ranges. When to use: We should use histogram When we need the count of the variable in a plot.
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
plt.hist(a, bins= [0,20,40,60,80,100])
plt.title("histogram")
plt.show()
```

```python
b= np.array([100,101,105,115,201,202,205,209,258,245,269,354,325,326,321,325,324,412,425,415,412,435,456,487,489])
plt.hist(b, bins= [50,100,150,200,250,300,350,400,450])
plt.title("Histogram")
plt.show()
```
