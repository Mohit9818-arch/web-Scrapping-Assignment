# 2nd Project : IMDB’s Top rated 100 movies’ 

```python
from bs4 import BeautifulSoup
import requests
```

```python
page= requests.get('https://www.imdb.com/chart/top')
```

```python
page
```

Page Content

```python
soup = BeautifulSoup(page.content)
soup
```

Scrapping Movie Name(Rank & Title), IMDb Rating, Your rating, Year

```python
header= [] #empty list for store the
#Doubt 

for i in soup.find_all('td',class_="titleColumn"):
    header.append(i.text.split("\n(")[0])
    
header
```

```python
year= [] #empty list for store the
#Doubt 

for i in soup.find_all('span',class_="secondaryInfo"):
    year.append(i.text)
    
year
```

```python
rating= [] #empty list for store the
#Doubt 

for i in soup.find_all('td',class_="ratingColumn imdbRating"):
    rating.append(i.text.split("n")[0])
    
rating
```

```python
#printing length
print(len(header),len(year),len(rating))
```

```python
#Making dataframe
import pandas as pd
df = pd.DataFrame({'Movie Name':header,'Year of Release':year,'Rating':rating})
df
```

```python
df.to_excel('Web_Scraping_Data2.xlsx',sheet_name='webscraping')
```

```python
df.to_csv('Web_Scraping_Data',index=False)
```
